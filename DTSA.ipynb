## **Data Types & Structures Basic Questions**

Here are **short answers** for each question:

1. **What are data structures, and why are they important?**
   
   Data structures are ways to organize and store data so it can be accessed and modified efficiently.

2. **Difference between mutable and immutable data types**
  
   Mutable types can be changed (e.g., list, dictionary). Immutable types cannot be changed after creation (e.g., string, tuple).

3. **Main differences between lists and tuples in Python**
  
   Lists are mutable and use `[ ]`; tuples are immutable and use `( )`.

4. **How dictionaries store data**
  
   Dictionaries store data as **key–value pairs** using hashing for fast lookup.

5. **Why use a set instead of a list**
  
   Sets automatically remove duplicates and allow faster membership checks.

6. **What is a string in Python, and how is it different from a list**
  
   A string is an immutable sequence of characters, while a list is a mutable collection of items.

7. **How tuples ensure data integrity**
  
   Tuples are immutable, so their values cannot be changed accidentally.

8. **What is a hash table, and relation to dictionaries**
  
   A hash table stores data using a hash function for fast access; Python dictionaries are implemented using hash tables.

9. **Can lists contain different data types?**
  
   Yes, lists can store multiple data types in the same list.

10. **Why strings are immutable in Python**
    
    Immutability improves memory efficiency,
    security, and performance.

11. **Advantages of dictionaries over lists**
    
    Dictionaries allow faster data retrieval using
    keys instead of searching through elements.

12. **How sets handle duplicate values**
    
    Sets automatically remove duplicates and only store unique elements.

13. **Scenario where tuple is better than list**
    
    When storing fixed data that should not change (e.g., coordinates `(x, y)`).

14. **How “in” works for lists vs dictionaries**
    
    In lists it searches values; in dictionaries it checks for keys.

15. **Can you modify tuple elements?**
    
    No, tuples are immutable, so their elements cannot be modified.

16. **What is a nested dictionary? Example use case**
    
    A dictionary inside another dictionary.
    Example: storing student details `{ "student1": {"age": 20, "marks": 85} }`.

17. **Time complexity of accessing dictionary elements**
    
    Average case is **O(1)** (constant time).

18. **When are lists preferred over dictionaries?**
    
    When order matters and data does not need key-value mapping.

19. **Why dictionaries are considered unordered (conceptually)**
    
    Because items are accessed by keys, not position, so index-based ordering is not the main focus.

20. **Difference between list and dictionary in data retrieval**
    
    Lists retrieve data using **index positions**, while dictionaries retrieve data using **keys**.
