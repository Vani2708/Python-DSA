## Functions Theory Questions**

1. Difference between a Function and a Method in Python
A function is a block of reusable code defined using the def keyword and can be called independently.
A method is a function that is associated with an object or class and is called using that object.

Example
Function
def greet(name):
    return "Hello " + name
print(greet("Rahul"))

Method
text = "hello world"
print(text.upper())
Here:
greet() → function
upper() → method of the string object text

2. Function Arguments and Parameters
Parameter: Variable defined in the function declaration.
Argument: Actual value passed to the function when calling it.

Example
def add(a, b):   # a and b are parameters
    return a + b
print(add(5, 3))   # 5 and 3 are arguments
Types of arguments:
Positional arguments
Keyword arguments
Default arguments
Variable-length arguments (*args, **kwargs)

3. Different Ways to Define and Call a Function in Python
1. Simple Function
def greet():
    print("Hello")
greet()

2. Function with Parameters
def square(x):
    return x * x
print(square(4))

3. Function with Default Arguments
def greet(name="Guest"):
    print("Hello", name)
greet()
greet("Anita")

4. Function with Variable Arguments
def total(*numbers):
    return sum(numbers)
print(total(2,3,4))

4. Purpose of the return Statement
The return statement sends a value from a function back to the caller.
Without return, the function returns None.
Example
def multiply(a, b):
    return a * b
result = multiply(4, 5)
print(result)
Output:
20

5. Iterators vs Iterables in Python
Iterable
An object that can be looped over using a for loop.

Examples: lists, tuples, strings, sets, dictionaries.
Iterator
An object that keeps track of iteration and produces the next value using next().

Example
numbers = [1,2,3]     # iterable
it = iter(numbers)    # iterator
print(next(it))
print(next(it))
print(next(it))

Difference:
Iterable	Iterator
Can be looped over	Produces items one at a time
Uses __iter__()	Uses __next__()

6. Generators in Python
A generator is a special type of function that produces values one at a time using yield instead of return.
Generators save memory because they do not store the entire sequence.

Example
def count_up(n):
    for i in range(n):
        yield i
gen = count_up(5)
for num in gen:
    print(num)
Output
0
1
2
3
4

7. Advantages of Generators
Memory Efficient – values generated one at a time
Faster execution for large datasets
Lazy evaluation (values produced only when needed)

Example
Generator:
def squares(n):
    for i in range(n):
        yield i*i
Instead of storing all values in a list:
[i*i for i in range(n)]
Generators are better for large n.

8. Lambda Function in Python
A lambda function is a small anonymous function defined using the lambda keyword
It can have multiple inputs but only one expression.

Syntax
lambda arguments : expression
Example
square = lambda x: x*x
print(square(5))

Output
25
Commonly used with map(), filter(), and sorted().

9. Purpose and Usage of map() Function
The map() function applies a function to every item of an iterable.
Syntax
map(function, iterable)
Example
numbers = [1,2,3,4]
squares = list(map(lambda x: x*x, numbers))
print(squares)
Output
[1, 4, 9, 16]

10. Difference Between map(), filter(), and reduce()
Function	Purpose
map()	Applies a function to each element
filter()	Selects elements based on condition
reduce()	Reduces iterable to a single value
Example
map()
nums = [1,2,3]
result = list(map(lambda x: x*2, nums))
Output
[2,4,6]
filter()
nums = [1,2,3,4,5]
even = list(filter(lambda x: x%2==0, nums))
Output
[2,4]
reduce()
from functools import reduce
nums = [1,2,3,4]
sum_all = reduce(lambda a,b: a+b, nums)
print(sum_all)
Output
10

```python
def sum_even(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

nums = [1, 2, 3, 4, 5, 6]
print("Sum of even numbers:", sum_even(nums))
```

```python
def is_prime(n):
    if n <= 1 or n > 200:
        return False

    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

num = 37
if is_prime(num):
    print(num, "is Prime")
else:
    print(num, "is Not Prime")
```

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value

fib = Fibonacci(6)

for num in fib:
    print(num)
```

```python
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

for value in powers_of_two(5):
    print(value)
```

```python
def read_file(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()

for line in read_file("sample.txt"):
    print(line)
```

```python
data = [(1, 5), (2, 3), (4, 1), (3, 2)]

sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)
```
