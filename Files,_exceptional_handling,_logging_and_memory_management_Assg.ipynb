# **Files, exceptional handling, logging and memory management Basic Questions**

1. **Interpreted vs Compiled Languages**

   * **Compiled:** Code is converted to machine code before execution (e.g., C).
   * **Interpreted:** Code is executed line by line (e.g., Python).

2. **Exception Handling in Python**
   A way to **handle runtime errors** using `try`, `except`, `else`, and `finally`.

3. **Purpose of `finally` Block**
   Code in `finally` **always executes**, whether an exception occurs or not.

4. **Logging in Python**
   A way to **record events, errors, and messages** during program execution.

5. **Significance of `__del__` Method**
   A **destructor method** called when an object is deleted.

6. **`import` vs `from ... import`**

   * `import module` → use `module.function()`
   * `from module import function` → use `function()` directly.

7. **Handling Multiple Exceptions**

```python
try:
    pass
except (ValueError, TypeError):
    pass
```

8. **Purpose of `with` Statement (Files)**
   Automatically **opens and closes files safely**.

9. **Multithreading vs Multiprocessing**

   * **Multithreading:** Multiple threads in one process.
   * **Multiprocessing:** Multiple independent processes.

10. **Advantages of Logging**

* Debugging
* Error tracking
* Monitoring program behavior

11. **Memory Management in Python**
    Process of **allocating and freeing memory automatically**.

12. **Basic Steps of Exception Handling**

* `try` → risky code
* `except` → handle error
* `else` → run if no error
* `finally` → always run

13. **Importance of Memory Management**
    Prevents **memory leaks and improves performance**.

14. **Role of `try` and `except`**

* `try`: Contains code that may cause error.
* `except`: Handles the error.

15. **Python Garbage Collection**
    Automatically **removes unused objects from memory**.

16. **Purpose of `else` in Exception Handling**
    Runs **if no exception occurs in try block**.

17. **Common Logging Levels**

* DEBUG
* INFO
* WARNING
* ERROR
* CRITICAL

18. **`os.fork()` vs `multiprocessing`**

* `os.fork()` creates a **child process (Unix only)**.
* `multiprocessing` is **cross-platform process creation**.

19. **Importance of Closing a File**
    Frees **system resources and saves changes**.

20. **`file.read()` vs `file.readline()`**

* `read()` → reads entire file.
* `readline()` → reads one line at a time.

21. **Logging Module Purpose**
    Used to **track events, debug information, and errors**.

22. **`os` Module in File Handling**
    Used for **file and directory operations** like create, delete, rename.

23. **Challenges in Memory Management**

* Memory leaks
* Fragmentation
* High memory usage

24. **Raise Exception Manually**

```python
raise ValueError("Invalid input")
```

25. **Importance of Multithreading**
    Improves **performance and responsiveness** in tasks like I/O operations.

````python
import logging
logging.basicConfig(filename="error.log", level=logging.ERROR)

try:
    a = 10/0
except ZeroDivisionError:
    logging.error("Division by zero occurred")
````
