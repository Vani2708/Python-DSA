### **Python Basic Questions**

1. What is Python, and why is it popular?

Ans: Python is a high-level, easy-to-read programming language. It’s popular because it is simple, versatile, and used in web development, AI, data science, and automation.

2. What is an interpreter in Python?

Ans: An interpreter executes Python code line by line instead of compiling the entire program at once.

3. What are pre-defined keywords in Python?

Ans: Keywords are reserved words in Python with special meanings (e.g., if, else, while, for, True, False, class).

4. Can keywords be used as variable names?

Ans: No. Keywords cannot be used as variable names because they already have predefined meanings.

5. What is mutability in Python?

Ans: Mutability means whether an object’s value can be changed after creation.

6. Why are lists mutable, but tuples are immutable?

Ans: Lists are designed for modifying data, while tuples are meant for fixed, unchangeable data.

7. What is the difference between “==” and “is” operators in Python?

Ans: Difference between == and is operators
== checks value equality
is checks object identity (same memory location)

8. What are logical operators in Python?

Ans: They combine conditional statements:
and, or, not.

9. What is type casting in Python?

Ans: Type casting means converting one data type into another (e.g., int("5")).

10. What is the difference between implicit and explicit type casting?

Ans: Difference between implicit and explicit type casting

Implicit: Python converts automatically.

Explicit: Programmer converts using functions like int(), float(), str().

11. What is the purpose of conditional statements in Python?

Ans: Purpose of conditional statements in Python
They allow programs to make decisions based on conditions (if, elif, else).

12. How does the elif statement work?
Ans: elif checks another condition if the previous if condition is false.

13. What is the difference between for and while loops?

Ans: Difference between for and while loops

for: Used when the number of iterations is known.

while: Used when the loop runs until a condition becomes false.

14. Describe a scenario where a while loop is more suitable than a for loop

Ans: Scenario where a while loop is better than a for loop
When the number of iterations is unknown, e.g., running a loop until a user enters the correct password.
